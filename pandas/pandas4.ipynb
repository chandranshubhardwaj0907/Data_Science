```python
#  data cleaningg and preprocessing script
import pandas as pd
```

```python
df = pd.read_csv('data_cleaning.csv')
df
```

```python
df.isnull()  # This will return a DataFrame of the same shape as df, with True for missing values and False for non-missing values
```

```python
df.isnull().sum() # This will give you the count of missing values in each column
```

```python
df.dropna()  # This will drop all rows with any missing values
```

```python
df.dropna(axis=1)  # This will drop all rows with any missing values
```

```python
df.fillna(0)  # This will fill all missing values with 0
```

```python
df["Age"].fillna(df["Age"].mean()) # used to fill missing values with the mean of the column
```

```python
df["Age"].fillna(df["Age"].std()) # used to fill missing values with the standard deviation of the column
```

```python
df["Age"].fillna(df["Age"].max())  # used to fill missing values with the maximum of the column
```

```python
df["Age"].fillna(df["Age"].min())  # used to fill missing values with the minimum of the column
```

```python

df["Age"].fillna(df["Age"].median()) # used to fill missing values with the median of the column
```

```python
df["Age"].fillna(df["Age"].mode()) # used to fill missing values with the mode of the column
```

```python
df.fillna(method = "ffill") # used to fill missing values by propagating the previous valid observation forward.
```

```python
df.fillna(method = "bfill")  # used to fill missing values by propagating the next valid observation backward.
```

```python
# find duplicates
df.duplicated()  # This will return a Series of boolean values indicating whether each row is a duplicate
```

```python
df.duplicated().sum()  # This will return the count of duplicate rows in the DataFrame
```

```python
df.drop_duplicates()  # This will drop all duplicate rows from the DataFrame
```

```python
df.duplicated(subset =['Name', 'Age'])  # This will check for duplicates based on the 'Name' and 'Age' columns
```

```python
# string methods 

df["Name"].str.lower()  # This will convert all strings in the 'Name' column to lowercase
df["Name"].str.upper()  # This will convert all strings in the 'Name' column to uppercase
df["Name"].str.title()  # This will convert all strings in the 'Name' column to title case
```
